```python
# 사용자가 특정 아이템을 선호하는 경우, 그 아이템과 비슷한 콘텐츠를 가진 다른 아이템을 추천해주는 방식
# 사용자가 a 영화에 높은 평점을 주었을때 그 영화의 특징을 기반으로 다른 영화를 추천함

#데이터 설명
#영화 출시 전 어떤 영화들이 좋은 평가 받을지 / 상업적인 성공을 거둘지 예측하기
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./tmdb_5000_movies.csv')
df.head(10)
```

```python
#사용할 컬럼 정리
# (순서대로) 평점 평균, 평점 카운트, 인기도
```

```python
data = df[['id', 'genres', 'vote_average', 'vote_count', 'popularity', 'title', 'keywords', 
           'overview']]
data.head()
```

```python
#평점 전처리
m = data['vote_count'].quantile(0.9)
print(m)
data = data.loc[data['vote_count']>=m]

c = data['vote_average'].mean()
```

```python
def weight_rating (x, m=m, c=c):
  v = x['vote_count']
  r = x ['vote_average']
  return (v / (v+m) *r ) + (m / (v+m) *c)
```

```python
data['score'] = data.apply(weight_rating, axis = 1)
```

```python
data.head()
```

```python
data.shape
```

```python
# 콘텐츠 기반 필터링을 위한 컬럼 추출
# 비슷함의 기준을 장르와 키워드로 잡기.
# 장르와 키워드 값은 list 안에 dict 이 포함된 구조. string 으로 변환 필요
# ast 패키지의 literal_eval 로 변환시킨다
```

```python
data[['genres', 'keywords']].head()
```

```python
import ast as ast
```

```python
data['genres'] = data['genres'].apply(ast.literal_eval)
data['keywords'] = data['keywords'].apply(ast.literal_eval)
```

```python
data[['genres', 'keywords']].head()
```

```python
data['genres'] = data['genres'].apply(lambda x :[d ['name'] for d in x]).apply(lambda x: " ".join(x))
data['keywords'] = data['keywords'].apply(lambda x :[d ['name'] for d in x]).apply(lambda x: " ".join(x))
```

```python
data['keywords'] = data['keywords'].apply(lambda x :[d ['name'] for d in x]).apply(lambda x: " ".join(x))
```

```python
# 콘텐츠 기반 필터링 추천
# 장르 벡터화
# 장르를 기준으로 필터링할 경우 -> 장르가 띄어쓰기가 구분된 한 문자열
# 문자열을 숫자로 바꿔 벡터화, CountVectorizer 사용
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer : 문서 집합에서 단어 토큰을 생성하고 각 단어의 수를 세어 BOW 인코딩 벡터를 만든다.
```

```python
counter_vector = CountVectorizer(ngram_range = (1,3))
c_vector_genres = counter_vector.fit_transform(data['genres'])
```

```python
c_vector_genres.shape
```

```python

corpus = [
    'This is the first document.',
    'This is the second second document.',
    'And the third one.',
    'Is this the first document?',
    'The last document?',
]
vect = CountVectorizer(ngram_range=(1, 3)).fit(corpus)
vect.fit(corpus)
vect.vocabulary_
```

```python
# 유사도값 추출
from sklearn.metrics.pairwise import cosine_similarity
similarity_genre = cosine_similarity(c_vector_genres, c_vector_genres).argsort()[:, ::-1]
print(similarity_genre)
```

```python
similarity_genre.shape
similarity_genre
```

```python
# 장르기반의 유사도를 기준으로 영화를 추천해주는 사용자 함수
def rc_movie_list(df, movie_title, top = 30):
  #특정 영화정보 뽑아내기
  target_mv_index = df[df['title'] == movie_title].index.values
  #타겟영화와 비슷한 코사인 유사도값
  sim_index = similarity_genre[target_mv_index, : top].reshape(-1)
  #본인제외
  sim_index = sim_index[sim_index!= target_mv_index]
  #추천결과 새로운 DF 생성, 평균평점(score) 으로 정렬
  result = df.iloc[sim_index].sort_values('score', ascending = False)[:10]
  return result
```

```python
rc_movie_list(data, movie_title = 'Avatar')
```
